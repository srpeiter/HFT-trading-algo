```python
import numpy as np
from matplotlib import pyplot as plt
import pandas as pd
from scipy.ndimage.interpolation import shift
```

```python
# binomial options pricing method
n = 15  # periods
T = 0.25 # maturity 3 months
S0 = 100 # price at t0
K = 110 # strike price
r = 1+ 2/100. #risk-free interest rate
sigma = 30/100. # volatility
c = 1/100. #dividend yield
u = 1.0395 
d = 1./u
q = (r-d)/(u-d)

```

```python
# u=1.06
# d = 1./u
# q = (r-d)/(u-d)
1/r *(16.167*q + (1-q)*6.76)
```

```python
pricetree = np.zeros((n,n))
for i in range(n):
    S_n = S0*u**i
    pricetree += np.diag(S_n*d**np.arange(0,n-i),k=i)
    
```

```python
C_n = np.zeros((n,n))
zeros = np.zeros((n))
C_n[:,n-1] = np.max(np.array([pricetree[:,(n-1)] - K,zeros]),axis=0)
for j in range(n-1):
    C_n_aux = shift(C_n[:,n-1-j],-1,cval=0)
    C_n[:,n-2-j] = np.round(1/r *(C_n[:,n-1-j]*q + (1-q)*C_n_aux),4)
    
    print(j)
    
    
```
