### Obtaining data from other stocks

```python
def get_stock_data(tickers, startdate, enddate):
    
    split_data = []
    data = pdr.get_data_yahoo(tickers,start=startdate, end=enddate)
    
    for tick in tickers:
        
        split_data.append(data.xs(tick, level='Symbols', axis=1))
    
    multi_d_data = pd.concat(split_data,keys=tickers,names=\
          ['Ticker','Date'])
        
    return multi_d_data
```

### Building a trading strategy

There are many trading strategies. These can be categorized in a momentum trading strategy and mean reversion strategy. The first one is a strategy based on the believe that the movement stock is to continue in one or the other direction. The latter is a strategy where one believe that the price of stock will settle to it mean value.
Let's start with the most simple trading strategy: the moving average crossover (momentum strategy)

```python
# defining 2 moving average: a short and long one 
# when short one exceeds the long one, a buy signal is generated.
# Vice versa, a sell signal is generated

short_window = 40
long_window = 100

# getting date index
date_ind = aapl.index

signals = pd.DataFrame(0.0,index=date_ind,columns = ['signal'])

# creating moving averages
signals['Close'] = aapl['Close']
signals['short_mavg'] = aapl['Close'].rolling(window=short_window,min_periods=1, center=False).mean()
signals['long_mavg'] = aapl['Close'].rolling(window=long_window,min_periods=1, center=False).mean()

# generating trading signals
cond = signals['short_mavg'] > signals['long_mavg']
trade_sig = a.apply(lambda x: 1 if x==True else 0)
signals['signal'] = trade_sig
signals.iloc[:short_window]['signal']=0
signals['positions'] = signals['signal'].diff()
```

```python
fig, ax1 = plt.subplots(figsize=(9,9))
signals[['Close','short_mavg','long_mavg']].plot(ax=ax1)

ax1.plot(signals.loc[signals.positions == 1.0].index, 
         signals.short_mavg[signals.positions == 1.0],
         '^', markersize=10, color='m')

ax1.plot(signals.loc[signals.positions == -1.0].index, 
         signals.short_mavg[signals.positions == -1.0],
         'v', markersize=10, color='k')
```

### Backtesting strategy

```python
initial_capital = 1000
positions = pd.DataFrame(index=date_ind,columns=['AAPL']).fillna(0)
portfolio = pd.DataFrame(index=date_ind,columns = ['val'])
# buy stocks
positions.AAPL = signals.signal*100

# buy and sell signals 
pos_diff = positions.diff()

# # market value of open positions
portfolio['holding'] = positions.AAPL * aapl['Adj Close']

portfolio['cash'] = initial_capital - (pos_diff.AAPL*aapl['Adj Close']).cumsum()

portfolio['total'] = portfolio['cash'] + portfolio['holding']

# portfolio['total'] = portfolio['cash'] + portfolio['holding']
```
